```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

# Загрузка данных из CSV-файла
data = pd.read_csv('data.csv')
data = data.fillna(0)

# Определение количества компонент
n_components = 10

# Извлечение переменных
variables = data.iloc[:, 0:110]

# Выполнение PCA
pca = PCA(n_components=n_components)
pca.fit(variables)

# Получение факторов
factors = pca.transform(variables)

# Расчет eigenvalue, difference, proportion и cumulative
eigenvalues = pca.explained_variance_
differences = np.diff(eigenvalues)
proportions = pca.explained_variance_ratio_
cumulative = np.cumsum(proportions)

# Создание объекта DataFrame для сохранения результатов в файл CSV
results_df = pd.DataFrame({'Component': range(1, n_components+1),
                           'Eigenvalue': eigenvalues,
                           'Difference': np.append(differences, np.nan),
                           'Proportion': proportions,
                           'Cumulative': cumulative})

# Сохранение результатов в файл CSV
results_df.to_csv('results.csv', index=False)

```

```python
import pandas as pd
from sklearn.decomposition import PCA

# Загрузка данных из файла CSV
data = pd.read_csv('data.csv')
data = data.fillna(0)

# Выбор столбцов для анализа
variables = data.iloc[:, 0:110]

# Создание объекта PCA с числом компонент равным 10
pca = PCA(n_components=10)

# Применение PCA к выбранным переменным
pca.fit(variables)

# Получение факторов
factors = pca.transform(variables)


# Получение факторных нагрузок
loadings = pca.components_.T * \
           np.sqrt(pca.explained_variance_)



# Создание списка имен переменных
variable_names = list(variables.columns)

# Создание списка имен факторов
factor_names = ['Factor 1', 'Factor 2', 'Factor 3', 'Factor 4', 'Factor 5', 
                'Factor 6', 'Factor 7', 'Factor 8', 'Factor 9', 'Factor 10']

# Создание DataFrame с факторными нагрузками и именами переменных
loadings_df = pd.DataFrame(loadings, columns=factor_names, index=variable_names)

# Вывод факторных нагрузок
# Сохранение таблицы в файл CSV
loadings_df.to_csv('loadings.csv', index_label='Variable')
```
